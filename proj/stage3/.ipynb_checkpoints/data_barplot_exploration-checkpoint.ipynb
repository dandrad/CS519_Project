```python
from load_and_reduce import load_and_reduce
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np

goodgreen = "#18453B"
goodlightblue = "#32dde3"
goodpurple = "#6d0585"
burntred = "#8a0303"
```

```python
intake_path = "Austin_Animal_Center_Intakes.csv"
outcome_path = "Austin_Animal_Center_Outcomes.csv"

X, y = load_and_reduce(intake_path, outcome_path, coding = None, scale = False, DimRed = "None")
X.head()
```

```python
def make_bar_plot(X,Y, color = goodgreen, xlab = "xlabel", ylab = "ylabel", title = "title", savename = "None", freq_text_center = 0.03):
    
    fig, ax = plt.subplots(figsize = (16,8))


    plt.bar(X,Y, color = color, linewidth = 4,edgecolor = "black")

    plt.xlabel(xlab,fontsize = 20, labelpad = 20)
    plt.ylabel(ylab, fontsize = 20, labelpad = 20)
    plt.title(title, fontsize = 26)



    plt.xticks(fontsize = 15)
    plt.yticks(fontsize = 15)

    plt.tick_params(length = 6, width = 3)

    for axis in ['top','bottom','left','right']:
        ax.spines[axis].set_linewidth(3)

    for i, v in enumerate(Y):
        ax.text(i - freq_text_center * len(str(v)), 800, str(v), color='black', fontweight='bold',fontsize = 18)

    plt.tight_layout()
    
    if(savename != "None"):
        
        plt.savefig(savename, dpi = 400)

```

```python
index_intake_type, counts_intake_type = np.unique(X["Intake Type"],return_counts=True)

color = goodgreen
xlab = "Intake Type"
ylab = "Count"
title = "Distribution of Intake Type"
savename = "./figures/intake_type_bar.png"
freq_text_center = 0.03


make_bar_plot(index_intake_type,counts_intake_type, color = color, xlab = xlab, ylab = ylab, title = title, savename = savename, freq_text_center = freq_text_center)
```
