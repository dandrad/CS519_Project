```python
from load_and_reduce import load_and_reduce


from sklearn.ensemble import GradientBoostingRegressor
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
```

```python
intake_path = "Austin_Animal_Center_Intakes.csv"
outcome_path = "Austin_Animal_Center_Outcomes.csv"

X, y = load_and_reduce(intake_path, outcome_path, coding = "onehot", scale = True)

X.head()
```

```python
X_copy = X.copy()
y_copy = y.copy()

nattributes = X.shape[1]
ninstances = X.shape[0]
```

```python
def GetModel():
    
    model = Sequential()
    
    model.add(Dense(nattributes, input_dim=nattributes,activation='relu')) 
    #model.add(Dense(64*4, activation='relu'))
    #model.add(Dense(64*4, activation='relu'))
    #model.add(Dense(64*2, activation='relu'))
    #model.add(Dense(64, activation='relu')) 
    model.add(Dense(64/2,activation='relu')) 
    model.add(Dense(16,activation='relu'))
    model.add(Dense(1,activation='linear'))
    model.compile(loss="mean_squared_error", optimizer="adam") 
    
    return model
```

```python
regressor = KerasRegressor(build_fn=GetModel,epochs=50,batch_size=50)
regressor.fit(X,y) 
```
